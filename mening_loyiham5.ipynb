```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
fig, axs = plt.subplots(2,2,figsize=(18,9))
kasallanish = pd.read_csv ('Aholining_kasallanishi.csv')
kasallanish
df = pd.DataFrame(kasallanish)
x = list(df.iloc[:, 0])
plt.xticks(rotation = 90)
y = list(df.iloc[:, 4])
plt.yticks(rotation = 90)
z = list(df.iloc[:, 5])
w = list(df.iloc[:, 6])
k = list(df.iloc[:, 7])
fig.suptitle('Ozbekiston aholisining yuqumli kasalliklar bilan kasallanish korsatkichlari')
plt.xticks(rotation = 90)
axs[0,0].plot(x, y,'tab:red')
axs[0, 0].set_title('Ichterlama va paratiflar, mingta holat')
axs[0,1].plot(x, z, 'tab:purple')
axs[0, 1].set_title('Gripp va ORVI, mingta holat')
axs[1,0].plot(x, w, 'tab:brown')
axs[1, 0].set_title('Virusli gepatit, mingta holat')
axs[1,1].plot(x, k)
axs[1,1].set_title('Skarlatina, mingta holat')

fig.autofmt_xdate(rotation=90)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
fig, axs = plt.subplots(2,2,figsize=(18,9))
emlash = pd.read_csv ('bolalarni-emlash.csv')

df = pd.DataFrame(emlash)
plt.xticks(rotation = 90)
x = list(df.iloc[:, 0])
y = list(df.iloc[:, 2])
z = list(df.iloc[:, 3])
w = list(df.iloc[:, 4])
k = list(df.iloc[:, 5])

fig.suptitle('Ozbekistonda bir yoshgacha bolgan bolalarni emlash korsatkichi')
plt.xticks(rotation = 90)
axs[0,0].plot(x, y,'P')
plt.xticks(rotation = 90)
axs[0, 0].set_title('Kokyotalga qarshi emlash, foiz')
axs[0,1].plot(x, z, '+')
plt.xticks(rotation = 90)
axs[0, 1].set_title('Poliomielitga qarshi emlash, foiz')
axs[1,0].plot(x, w, 'x')
plt.xticks(rotation = 90)
axs[1, 0].set_title('Qoqsholga qarshi emlash, foiz')
axs[1,1].plot(x, k, 'X')
plt.xticks(rotation = 90)
axs[1,1].set_title('Qizamiqqa qarshi emlash, foiz')

fig.autofmt_xdate(rotation=90)
```
