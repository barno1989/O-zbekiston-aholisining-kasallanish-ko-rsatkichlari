```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
kasallanish = pd.read_csv ('Aholining_kasallanishi.csv')
kasallanish
df = pd.DataFrame(kasallanish)
  
X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])
  
plt.bar(X, Y, color='b')
plt.title("O'zbekiston aholisining o'tkir ichak infeksiyalari bilan kasallanishi darajasi")
plt.xlabel("Yillar")
plt.ylabel("O'tkir ichak infeksiyalari, ming")
  
plt.show()
```
