```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

with open('SS_korsatkichlari.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='.')
    for row in plots:
        x.append(int(row[10]))
        y.append(int(row[1]))

plt.plot(x,y, label='Loaded from file!')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sogliqni saqlash korsatkichlari')
plt.legend()
plt.show()
```
